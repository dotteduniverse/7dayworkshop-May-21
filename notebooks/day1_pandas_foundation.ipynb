## 1. Setup & First DataFrame

```python
# Import the pandas library (convention: pd)
import pandas as pd

# Create a simple DataFrame from a dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Age': [25, 30, 35, 28],
    'Salary': [50000, 60000, 70000, 52000]
}
df = pd.DataFrame(data)
print("Our first DataFrame:")
df
```

## 2. Reading a Real CSV Dataset

We'll use a sample dataset (bank customer data). If you don't have a local file, pandas can read from a URL.

```python
# Read from a public CSV (bank marketing data)
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
# Add column names because the raw file has none
columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigree', 'Age', 'Outcome']
df_bank = pd.read_csv(url, names=columns)
df_bank.head()  # First 5 rows
```

### Explore the data

- `.head()` – first rows
- `.info()` – data types and non-null counts
- `.describe()` – summary statistics
- `.shape` – dimensions (rows, columns)

```python
# Info: data types, missing values
df_bank.info()
```

```python
# Describe: statistics for numerical columns
df_bank.describe()
```

```python
# Shape: (rows, columns)
print(f"Dataset has {df_bank.shape[0]} rows and {df_bank.shape[1]} columns.")
```

## 3. Selecting Data

### `.loc[]` – label-based selection (row labels, column names)

```python
# Select first 3 rows and columns 'Glucose' and 'BMI'
df_bank.loc[0:2, ['Glucose', 'BMI']]
```

### `.iloc[]` – position-based selection (integer indices)

```python
# Select rows 0,1,2 and columns 1 and 5 (Glucose and BMI)
df_bank.iloc[0:3, [1, 5]]
```

### Boolean filtering (conditions)

```python
# Filter rows where Age > 50
df_bank[df_bank['Age'] > 50].head()
```

```python
# Multiple conditions: Age > 50 AND Glucose > 150
filtered = df_bank[(df_bank['Age'] > 50) & (df_bank['Glucose'] > 150)]
print(f"Number of matching rows: {len(filtered)}")
filtered.head()
```

## 4. Data Manipulation

### Rename columns

```python
# Rename 'DiabetesPedigree' to 'Pedigree'
df_renamed = df_bank.rename(columns={'DiabetesPedigree': 'Pedigree'})
df_renamed.columns
```

### Sort values

```python
# Sort by Age descending
df_bank.sort_values('Age', ascending=False).head()
```

### Handle missing values (fillna)

Some datasets have zeros where missing should be (e.g., BMI=0 is impossible). Let's create a copy with some NaN values for practice.

```python
# Create a copy and set first 5 rows of BMI to NaN
df_missing = df_bank.copy()
df_missing.loc[0:4, 'BMI'] = None
print("Missing values per column:")
df_missing.isnull().sum()
```

```python
# Fill NaN with the mean of BMI (excluding NaNs)
mean_bmi = df_missing['BMI'].mean()
df_missing['BMI'] = df_missing['BMI'].fillna(mean_bmi)
print(f"Filled NaNs with mean BMI: {mean_bmi:.2f}")
df_missing[['BMI']].head()
```

### Drop rows with any missing values

```python
# Recreate missing and drop rows with any NaN
df_missing2 = df_bank.copy()
df_missing2.loc[0:4, 'BMI'] = None
df_cleaned = df_missing2.dropna()
print(f"Original rows: {len(df_missing2)}, After dropna: {len(df_cleaned)}")
```

## 5. Grouping & Aggregation

Use `.groupby()` to split data into groups and apply a function (like mean).

```python
# Group by 'Outcome' (0 = no diabetes, 1 = diabetes) and calculate mean of Glucose and Age
grouped = df_bank.groupby('Outcome')[['Glucose', 'Age']].mean()
grouped
```

```python
# More complex: multiple aggregations
grouped2 = df_bank.groupby('Outcome').agg({
    'Glucose': ['mean', 'std'],
    'Age': ['min', 'max']
})
grouped2
```

## 6. Practical Exercise (Test Yourself)

Using `df_bank`, complete the following tasks:

1. Filter rows where `BloodPressure` is less than 60.
2. Create a new column `AgeGroup` that categorises Age into `Young` (<30), `Middle` (30-50), `Senior` (>50).
3. Group by `AgeGroup` and calculate the average `BMI` for each group.
4. Sort the dataset by `Glucose` descending and show the top 5 rows.

*Try to write the code in the cell below without looking at solutions.*

```python
# Your code here

```

### Solution (Check after trying)

```python
# 1. Low blood pressure
low_bp = df_bank[df_bank['BloodPressure'] < 60]
print(f"Rows with BP < 60: {len(low_bp)}")

# 2. Create AgeGroup column
def age_group(age):
    if age < 30:
        return 'Young'
    elif age <= 50:
        return 'Middle'
    else:
        return 'Senior'

df_bank['AgeGroup'] = df_bank['Age'].apply(age_group)

# 3. Group by AgeGroup and mean BMI
bmi_by_agegroup = df_bank.groupby('AgeGroup')['BMI'].mean()
print("\nAverage BMI by AgeGroup:")
print(bmi_by_agegroup)

# 4. Sort by Glucose descending
sorted_glucose = df_bank.sort_values('Glucose', ascending=False)
print("\nTop 5 rows by Glucose:")
sorted_glucose[['Glucose', 'Age', 'AgeGroup']].head()
```
